```python
from Bio import Entrez

def search_pubmed(gene_name):
    # Perform a search in PubMed using Entrez esearch
    handle = Entrez.esearch(db="pubmed", term=gene_name, retmax=100)
    record = Entrez.read(handle)
    handle.close()
    
    # Get the list of PubMed IDs (PMIDs) from the search result
    id_list = record["IdList"]
    return id_list

def fetch_details(id_list):
    ids = ",".join(id_list)
    handle = Entrez.efetch(db="pubmed", id=ids, rettype="xml", retmode="text")
    records = Entrez.read(handle)
    handle.close()
    return records

def extract_information(records):
    articles_info = []
    for article in records['PubmedArticle']:
        pmid = article['MedlineCitation']['PMID']
        title = article['MedlineCitation']['Article']['ArticleTitle']
        year = article['MedlineCitation']['Article']['Journal']['JournalIssue']['PubDate'].get('Year', 'Not Available')
        articles_info.append({'PMID': pmid, 'Title': title, 'Year': year})
    return articles_info

gene_name = "BRCA1"  # Replace with your gene of interest
id_list = search_pubmed(gene_name)
records = fetch_details(id_list)
articles_info = extract_information(records)

for article in articles_info:
    print(f"PMID: {article['PMID']}, Title: {article['Title']}, Year: {article['Year']}")
```
